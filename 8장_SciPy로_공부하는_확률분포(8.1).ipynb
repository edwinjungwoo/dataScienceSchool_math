# 사이파이를 이용한 확률분포 분석

```python
import scipy as sp
import scipy.stats
```

```python
rv = sp.stats.norm()
```

```python
rv = sp.stats.norm(loc=1, scale=2)
# 기댓값이 1이고 표준 편차가 2인 정규분포
```

```python
import matplotlib.pyplot as plt
import numpy as np
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title("pdf")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.grid()
plt.show()
```

```python
rv = sp.stats.norm(loc = 0, scale = 0.1)
xx = np.linspace(-1, 1, 1000)
pdf = rv.pdf(xx)

plt.plot(xx, pdf)
plt.grid()
plt.show()
```

```python
rv.pdf(0)
```

```python
rv = sp.stats.norm(loc=1, scale=2)
# 기댓값이 1이고 표준 편차가 2인 정규분포

xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title("cdf")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.grid()
plt.show()
```

```python
rv.rvs(size=(3, 5), random_state=0)
```

```python
import seaborn as sns
sns.distplot(rv.rvs(size=10000, random_state=0))
plt.title("random sample")
plt.xlabel("samples")
plt.ylabel("count")
plt.xlim(-8, 8)
plt.grid()
plt.show()
```

```python
np.random.seed(0)
x = sp.stats.norm().rvs(1000)
print('표본평균 :', x.mean())
print('비편향 표본분산 :', x.std(ddof=1))
```

```python
rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform()

np.random.seed(0)
N = 50000
x_1 = rv1.rvs(N)
x_2 = rv2.rvs(N)
x_3 = x_1 + x_2

plt.figure(figsize=(12, 5))

plt.subplot(131)
sns.distplot(x_1, kde=False)
#plt.title("균일분포")
#plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(132)
sns.distplot(x_2, kde=False)
#plt.title("균일분포")
#plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(133)
sns.distplot(x_3, kde=False)
#plt.title("균일분포 표본의 합의 분포")
#plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.show()
```

```python
x = sp.stats.uniform().rvs((10, 10000))
s = x.sum(axis=0)
sns.distplot(s)
plt.show()
```
